```python
#day 4: Pandas Data Mastery
import warnings
warnings.filterwarnings('ignore')


import pandas as pd
import numpy as np

print("day 4: Pandas Grouping, Maps & Data Cleaning")
```

```python
#1 Lets create sample data of store

store_data = {
    'salesperson' : ['alice', 'Mahir', 'Ali' ,'AAna' , 'Alice','Mahir' , 'Alice','Alice' ],
    'product' : ['laptop' , 'laptop' , 'mouse','keyboard','monitor', 'monitor' ,'mouse','printer'],
    'price' : [1000,1000,250,600,340,340,250,2000],
    'Quantity' :[1,2,3,1,3,4,1,2],
    'city' : ['NYC' ,'UN','CHICAGO','LA','NYC' ,'UN','NYC','LA']
    
    
}

df = pd.DataFrame(store_data)
print ("The store sales data: \n ",df)
```

```python
#2 Data cleaning , lets fix the   inconsistent names
print("\n6. lets fix the in consistent name:")

print("Before cleaning, Salesperson sales counts:")
print(df['salesperson'].value_counts())

# fixing the issue
df['salesperson'] = df['salesperson'].str.title()

print("\n After cleaning, Salesperson sales counts:")
print(df['salesperson'].value_counts())
```

```python
#3 Grouping_ total sales by each person

sales_by_person = df.groupby('salesperson')['price'].sum()
print("total sales by salesperson")
print(sales_by_person)
```

```python
#4 Sorting , finding the best product

print("Most expensive product : \n" )
expensive_products = df.sort_values('price', ascending = False)
print(expensive_products[['product', 'price']])
```

```python
#5 Maping , catogrize products by price
def price_category(price):
    if price < 200:
        return "in budget"
    elif price < 500 :
        return "Mid_range"
    else:
        return "premium"

df['price_category'] = df['price'].map(price_category)
print(df ,"\n"* 4)
"or"
print(df[['product', 'price', 'price_category']])
```

```python
#6 Missing values ,handling the real data

print("\n Handling the missing data: ")
df_missing = df.copy()
df_missing.loc[2, 'price'] = None #create missing value

print("before cleaning : ")
print(df_missing[df_missing['price'].isnull()])


#Fill the missing value
df_clean = df_missing.fillna({'price' : df_missing['price'].mean()})
print("\nAfter cleaning:")
print(df_clean.iloc[2])
```

```python
# 5. Total sales of the stores
print("\n5. total revenue calculations:")
df['total_sale'] = df['price'] * df['Quantity']
print(df[['salesperson', 'product', 'total_sale']])

print(f"\n revenue of the total sales : ${df['total_sale'].sum()}")
print(" Day 4 Complete , All pandas concepts practiced!")
```
