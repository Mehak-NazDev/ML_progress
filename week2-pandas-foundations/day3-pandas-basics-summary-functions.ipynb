```python
#Day 3...keeping the connection alive

import pandas as pd

data ={'price' : [12,25,18,92,12], 'points' : [85,90,42,31,11] }
data_frame = pd.DataFrame(data)
print(data_frame)


print("quick summary of what we learned today : "
     "\n"f"mean prices {data_frame.price.mean()} "
    "\n"f"max points {data_frame.points.max()} ")

print(f"unique value in prices { data_frame.price.unique()} , unique value count {data_frame.price.nunique()}")
print(f"price median {data_frame.price.median()}")
```
