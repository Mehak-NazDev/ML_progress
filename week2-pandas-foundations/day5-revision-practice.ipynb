```python
#Day 5 : Quick pandas and Ml revision
# light practice

import pandas as pd
import numpy as np

print("Quick Revision , core concepts")
print("\n"*2)

#1. create simple data
data = {
    'student' : ['Ali','Sara','Ali','Ahmad','Sara'],
    'subject' : ['Math' , 'Science' ,'Science' , 'Math' , 'Math'],
'score' : [85,92,88,95,78]

}

df = pd.DataFrame(data)
print("students score" , df )
```

```python
print ("\n1. Average score by student: ")
avg_scores = df.groupby('student')['score'].mean()
print(avg_scores)
```

```python
#filter

print(" Top scores (Abopve 90): ","\n" *2) 
top_student = df[df['score']>90]
print(top_student)
```
