```python
#revising the day 1 concepts....recaping it

import pandas as pd

# making some dummy housing data (not real)
data = {
    'price': [250000, 350000, 450000, 300000, 400000],
    'bedrooms': [3, 4, 5, 3, 4],
    'area_sqft': [1500, 2000, 2500, 1800, 2200],
    'location': ['A', 'B', 'C', 'A', 'B']
}

df = pd.DataFrame(data)
df

# checking columns selection 
df['price']   # the single column

# multiple columns (select more than one column)
df[['price','bedrooms']]

# iloc = by index (position)
# trying few examples

df.iloc[0]     # first row 
df.iloc[:3]    # first 3 rows


# loc = label based indexing
# looks similar but works with labels

df.loc[0:2]    # rows 0 to 2 inclusive


# filtering stuff
df[df['price'] > 300000]

# more filters
df[df['bedrooms'] >= 4]

# combine both
df[(df['price'] > 300000) & (df['bedrooms'] >= 4)]



# modifying data
# new col: price per sqft

df['price_per_sqft'] = df['price'] / df['area_sqft']
df.head() # first few rows


# changing location names just for fun 
df.loc[df['location'] == 'A', 'location'] = 'Prime'
df
print("day 2 done....along with exams")
```
