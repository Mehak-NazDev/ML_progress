```python
#sales of small coffee shop in a day

import pandas as pd 
import numpy as np

data = { 
      'item' : [ 'Latte','Espresso', 'latte', 'Cappuccino','Latte'],
      'price' : [4.0,3.0,-5.0,3.5,0.0],
      'cups_ordered ': [2,1,1,500,3]    
}
df=pd.DataFrame(data)
df
```

```python

#code to remove any accidental space from the start or end of all column names 
df.columns = df.columns.str.strip()

#make item name lower case
df['item'] = df['item'].str.lower()

#remove accidental price entered as 0 or negative for latte
df.loc[(df['item']== 'latte') & (df['price'] <= 0) ,'price'] = 4.0
df
```

```python
#Removing the outlier(500 coffe cups in a day)
df.loc[df ['cups_ordered'] > 5  ,'cups_ordered'] = 5
df
```

```python
#Total sales of the day
total_revenue = (df['price'] * df['cups_ordered']).sum()
print(f'{total_revenue}$ is the total revenue of the day ')
```

```python

#grouping by clean lower case column

sales_report = df.groupby('item') ['cups_ordered'].sum().sort_values(ascending=False)
print(sales_report)
```

```python
#visualising the most selling and least selling item in a day
import matplotlib.pyplot as plt
import seaborn as sns


# create summary (using reset index so item becomes a column so we can plot)
report = df.groupby('item')['cups_ordered'].sum().reset_index()

#draw the bar chart
plt.figure(figsize=(8,5))
sns.barplot(data = report,x = 'item', y = 'cups_ordered', palette = 'viridis')


```
